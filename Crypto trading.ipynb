### Crypto trading margin calculator

The following calculator is used to calculate amount of money needed for margin trading of cryptocurrencies.

This calculator eleminates the risk of losing susbtaital amount of money in the long run by keeping the amount of risk (fixed % of deposit) under control.


```python
import pandas as pd
import numpy as np
```

```python
def long_profit(deposit = 10000, enter_price = 40000, stoploss = 0.07, risk = 0.01, target_sale_price = 44000):
    
    # When Margin = 1
    risk_amount = deposit * risk
    amount_to_invest = deposit / (stoploss / risk)
    amount_btc_buy = amount_to_invest / enter_price
    stoploss_price = enter_price - enter_price*stoploss
    acceptable_lost = enter_price * stoploss    #
    amount_usd_if_loss = amount_btc_buy * stoploss_price
    portion_of_dep = amount_to_invest / deposit
    
    #Potential profits
    potential_profit_at_5p = (amount_btc_buy*enter_price*1.05) - amount_to_invest
    potential_profit_at_10p =(amount_btc_buy*enter_price*1.1) - amount_to_invest
    potential_profit_at_20p =(amount_btc_buy*enter_price*1.20) - amount_to_invest
    
    if target_sale_price != 0:
        potential_profit_at_target = amount_btc_buy*target_sale_price - amount_to_invest
    
    
    # Margin = 5X
    #m5x_amount_to_invest = 4000 * 5
    own_money_to_invest_at_5x = amount_to_invest / 5
    m5x_amount_to_invest = own_money_to_invest_at_5x * 5
    money_borrowed_at_5x = m5x_amount_to_invest - own_money_to_invest_at_5x
    m5x_amount_btc_buy = m5x_amount_to_invest / enter_price
    stoploss_price = enter_price - enter_price*stoploss
    
    
    acceptable_lost = enter_price * stoploss    #
    amount_usd_if_loss = amount_btc_buy * stoploss_price
    portion_of_dep_5x = own_money_to_invest_at_5x / deposit
    m5x_liquidation_price = (m5x_amount_to_invest - own_money_to_invest_at_5x) / m5x_amount_btc_buy
    
    #Potential profits
    m5x_potential_profit_at_5p = (m5x_amount_btc_buy*enter_price*1.05) - m5x_amount_to_invest
    m5x_potential_profit_at_10p =(m5x_amount_btc_buy*enter_price*1.1) - m5x_amount_to_invest
    m5x_potential_profit_at_20p =(m5x_amount_btc_buy*enter_price*1.20) - m5x_amount_to_invest
    
    if target_sale_price != 0:
        m5x_potential_profit = m5x_amount_btc_buy*target_sale_price - m5x_amount_to_invest
    
    # Margin = 10X
    #m10x_amount_to_invest = 4000 * 10
    own_money_to_invest_at_10x = amount_to_invest / 10
    m10x_amount_to_invest = own_money_to_invest_at_10x * 10
    money_borrowed_at_10x = m10x_amount_to_invest - own_money_to_invest_at_10x
    
    m10x_amount_btc_buy = m10x_amount_to_invest / enter_price
    stoploss_price = enter_price - enter_price*stoploss
    acceptable_lost = enter_price * stoploss    #
    amount_usd_if_loss = amount_btc_buy * stoploss_price
    portion_of_dep_10x = own_money_to_invest_at_10x / deposit
    m10x_liquidation_price = (m10x_amount_to_invest - own_money_to_invest_at_10x) / m10x_amount_btc_buy
    
    #Potential profits
    m10x_potential_profit_at_5p = (m10x_amount_btc_buy*enter_price*1.05) - m10x_amount_to_invest
    m10x_potential_profit_at_10p =(m10x_amount_btc_buy*enter_price*1.1) - m10x_amount_to_invest
    m10x_potential_profit_at_20p =(m10x_amount_btc_buy*enter_price*1.20) - m10x_amount_to_invest
    
    if target_sale_price != 0:
        m10x_potential_profit = m10x_amount_btc_buy*target_sale_price - m10x_amount_to_invest

    print('NO MARGIN')
    print('................')
    print('DEPOSIT:' , "${:,.2f}".format(deposit))
    print("RISK: {0:.2f}%".format(risk * 100))
    print('ENTER BTC PRICE:' , "${:,.2f}".format(enter_price))
    print('STOPLOSS: ' + "{:.1%}".format(stoploss))
    print('STOPLOSS PRICE:' , "${:,.2f}".format(stoploss_price))
    print('AMOUNT BTC TO BUY:' , round(amount_btc_buy,5), '(' + str("${:,.2f}".format(amount_to_invest)) + ' TO SPEND)')
    print('PORTION OF DEPOSIT ' + "{:.2%}".format(portion_of_dep))
    print('AMOUNT $ IF SOLD AT STOPLOSS', round(amount_usd_if_loss,2), '(LOSS', str("${:,.2f}".format(amount_to_invest-amount_usd_if_loss)) + ')')
    print('................')
    print('PROFIT AT 5% PRICE INCREASE:', "${:,.2f}".format(potential_profit_at_5p))
    print('PROFIT AT 10% PRICE INCREASE:', "${:,.2f}".format(potential_profit_at_10p))
    print('PROFIT AT 20% PRICE INCREASE:', "${:,.2f}".format(potential_profit_at_20p))
    if target_sale_price != 0:
        print('PROFIT AT', target_sale_price, 'USDT TARGET PRICE:', "${:,.2f}".format(potential_profit_at_target))
    
    print('')
    print('5X MARGIN')
    print('................')
    print('ENTER BTC PRICE:' , "${:,.2f}".format(enter_price))
    print('STOPLOSS: ' + "{:.1%}".format(stoploss))
    print('STOPLOSS PRICE:' , "${:,.2f}".format(stoploss_price))
    print('AMOUNT BTC TO BUY:' , round(m5x_amount_btc_buy,5), '(' + str("${:,.2f}".format(m5x_amount_to_invest)) + ' /', "${:,.2f}".format(own_money_to_invest_at_5x), 'OWN)')
    print('PORTION OF DEPOSIT ' + "{:.2%}".format(portion_of_dep_5x))
    print('AMOUNT $ IF SOLD AT STOPLOSS', round(amount_usd_if_loss,2), '(LOSS', str("${:,.2f}".format(amount_to_invest-amount_usd_if_loss)) + ')')
    print('LIQUIDATION PRICE', "${:,.2f}".format(m5x_liquidation_price))
    print('................')
    print('PROFIT AT 5% PRICE INCREASE:', "${:,.2f}".format(m5x_potential_profit_at_5p))
    print('PROFIT AT 10% PRICE INCREASE:', "${:,.2f}".format(m5x_potential_profit_at_10p))
    print('PROFIT AT 20% PRICE INCREASE:', "${:,.2f}".format(m5x_potential_profit_at_20p))
    if target_sale_price != 0:
        print('PROFIT AT', target_sale_price, 'USDT TARGET PRICE:', "${:,.2f}".format(m5x_potential_profit))
        
    print('')
    print('10X MARGIN')
    print('................')
    print('ENTER BTC PRICE:' , "${:,.2f}".format(enter_price))
    print('STOPLOSS: ' + "{:.1%}".format(stoploss))
    print('STOPLOSS PRICE:' , "${:,.2f}".format(stoploss_price))
    print('AMOUNT BTC TO BUY:' , round(m10x_amount_btc_buy,5), '(' + str("${:,.2f}".format(m10x_amount_to_invest)) + ' /', "${:,.2f}".format(own_money_to_invest_at_10x), 'OWN)')
    #print('AMOUNT BTC TO BUY:' , round(m5x_amount_btc_buy,5), '(' + str(round(m5x_amount_to_invest,2)) +' USD TOTAL /', round(amount_to_invest,2), 'USD OWN)')
    print('PORTION OF DEPOSIT ' + "{:.2%}".format(portion_of_dep_10x))
    print('AMOUNT $ IF SOLD AT STOPLOSS', round(amount_usd_if_loss,2), '(LOSS', str("${:,.2f}".format(amount_to_invest-amount_usd_if_loss)) + ')')
    print('LIQUIDATION PRICE', "${:,.2f}".format(m10x_liquidation_price))
    print('................')
    print('PROFIT AT 5% PRICE INCREASE:', "${:,.2f}".format(m10x_potential_profit_at_5p))
    print('PROFIT AT 10% PRICE INCREASE:', "${:,.2f}".format(m10x_potential_profit_at_10p))
    print('PROFIT AT 20% PRICE INCREASE:', "${:,.2f}".format(m10x_potential_profit_at_20p))
    if target_sale_price != 0:
        print('PROFIT AT', target_sale_price, 'USDT TARGET PRICE:', "${:,.2f}".format(m10x_potential_profit))  

long_profit()
```
